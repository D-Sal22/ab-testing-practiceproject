## 1. Clean your Datasets

For the first part of your project, you will apply the same data cleaning steps to your "data/control.txt" and "data/treatment.txt" files. By the end of this step, you should have two lists of floats that match the numbers you have in each respective file.

You will implement your `clean_data` function in your `cleaner.py` module and import the module in the code-block below.

```python
# TODO: import all the packages that you will need for this notebook here
import cleaner
import statistics as stats
import matplotlib.pyplot as plt
```

```python
# TODO: read your text files into your code using file I/O
control_data = []
treatment_data = []

with open("./data/control.txt", "r") as f:
    control_data = f.readlines()

with open("./data/treatment.txt", "r") as f:
    treatment_data = f.readlines()

# TODO: clean both list objects that you created above using the 'clean_data' function
cleaned_control_data = cleaner.clean_data(control_data)
cleaned_treatment_data = cleaner.clean_data(treatment_data)

print("cleaned control list =", sorted(cleaned_control_data))
print("control_data type,", type( cleaned_control_data[0]))
print("cleaned control data count =", len(cleaned_control_data))
print("cleaned treatment list =", sorted(cleaned_treatment_data))
print("treatment_data type,", type(cleaned_treatment_data[0]))
print("cleaned treatment data count =", len(cleaned_treatment_data))
```

## 2. Explore your Datasets

Now that you have your list objects created and cleaned, let's explore the descriptive statistics of these two files and make subsequent observations on our data.

```python
# TODO: Calculate the standard deviation of both the control and treatment group
stdev_control_data = stats.stdev(cleaned_control_data)
stdev_treatment_data = stats.stdev(cleaned_treatment_data)

# TODO: print your findings
print(round(stdev_control_data, 2))
print(round(stdev_treatment_data, 2))
```

```python
# TODO: Calculate the mean of both the control and treatment group
avg_control_data = stats.mean(cleaned_control_data)
avg_treatment_data = stats.mean(cleaned_treatment_data)


# TODO: print your findings
print("control group average viewing time =", round(avg_control_data, 2))
print("treatment group average viewing time =", round(avg_treatment_data, 2))
```

```python
# TODO: Are there any outliers in your control group? (use 2 * stdev to find these outliers)
outliers_control = []
threshold = 2 
lower_outlier = avg_control_data - (threshold * stdev_control_data)
upper_outlier = avg_control_data + (threshold * stdev_control_data)

for x in cleaned_control_data:
    val = float(x)                                      #I dont have to define val here. I could just write x, but it helps me remember syntax and what I am actually doing.. 
    if val < lower_outlier or val > upper_outlier:
        outliers_control.append(val)

outliers_control.sort()
print(outliers_control)
print(f'The number of control group outliers are: {len(outliers_control)}')
```

```python
# TODO: Are there any outliers in your treatment group?
outliers_treatment = []
threshold = 2 
t_lower_outlier = avg_treatment_data - (threshold * stdev_treatment_data)
t_upper_outlier = avg_treatment_data + (threshold * stdev_treatment_data)

for x in cleaned_treatment_data:
    
    if x < t_lower_outlier or x > t_upper_outlier:
        outliers_treatment.append(x)

outliers_treatment.sort()
print(outliers_treatment)
print(f'The number of treatment group outliers are: {len(outliers_treatment)}')
```

```python
# TODO: create a data visualization to view differences in dispersion and mean
# I understood how to create the plot, but used copilot to explain how to add the coloring > "Create the plot with 'patch_artist=True' so we can fill the boxes"
mean_style = dict(marker='o', markerfacecolor='yellow', markeredgecolor='black')    #mean style

bp = plt.boxplot([cleaned_control_data, cleaned_treatment_data], 
                 labels=['Control', 'Treatment'], 
                 patch_artist=True,
                 showmeans=True,
                 meanprops=mean_style
                 )

# my notes: Access each box individually by its index [0] and [1]
# my notes: bp['boxes'] is just a list of the box shapes

# Set the first box (Control) to blue
control_box = bp['boxes'][0]
control_box.set_facecolor('blue')

# Set the second box (Treatment) to purple
treatment_box = bp['boxes'][1]
treatment_box.set_facecolor('purple')

#I asked co pilot for help creating a graph key
key_text = (
    "Key\n"
    "Blue Box: Control Group IQR\n"
    "Purple Box: Treatment Group IQR\n"
    "Red Line: Median Value\n"
    "Yellow Circle: Mean Value\n"
    "White Dots: Outliers"
)

# 3. Add the text box to the plot
# (.05, .1) are coordinates (x, y) relative to the plot axes (0 to 1)
plt.text(1,.30, key_text, transform=plt.gca().transAxes, fontsize=10,
         verticalalignment='top', bbox=dict(boxstyle='round', facecolor='white', alpha=0.5))



# Back to the original assignement. This portion I understood how to create

plt.title('Presentation Viewing Time\n Control Group vs Treatment Group Dispersion')
plt.xlabel("Expiremental Groups\n [Control Group = 50 word count per page, Treatment Group =15 word count per page]")
plt.ylabel("Viewing Time (minutes)")
plt.show()


```
