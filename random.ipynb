```python
import matplotlib.pyplot as plt

# Sample data
x = [1, 2, 3, 4]
y_sales = [10, 20, 25, 30]
y_costs = [5, 15, 12, 20]

# 1. Use the 'label' argument inside plot()
plt.plot(x, y_sales, label='Sales')
plt.plot(x, y_costs, label='Costs')

# 2. Call plt.legend() to display the labels
plt.legend()

# Display the plot
plt.title("Company Performance")
plt.xlabel("Month")
plt.ylabel("Amount")
plt.show()
```
